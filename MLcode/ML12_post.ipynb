```python
# 序列填充post

import tensorflow as tf     #导入所需库
s=[[1],[2,3],[4,5,6]]       #初始化填充前序列s
print("填充前序列为\n",s)  #显示填充前序列
a=tf.keras.preprocessing.sequence. pad_sequences(s, padding='post')
print("填充后序列为\n",a)  #显示填充后序列
```

```python
# 构建含嵌入层的Keras网络模型

import tensorflow as tf      #导入所需库
import numpy as np
model=tf.keras.Sequential()  #构建空的网络模型model
#创建嵌入层并添加到model中
embedding=tf.keras.layers.Embedding(output_dim=32, input_dim=100, input_length=400)
model.add(embedding)         #将嵌入层添加到model中
model.build(input_shape=(None, 400))
model.summary()              #显示深度学习网络的参数信息
#显示embeddings矩阵
e=np.array(embedding.get_weights())
print("embeddings矩阵的值\n",e)
#定义句子数据集
text="Deep learning is an important concept raised by the current sciences."
#分词
token=tf.keras.preprocessing.text.Tokenizer(num_words=100)
token.fit_on_texts(text)     #分词语句
#将词转换为整数序列
input=token.texts_to_sequences(text)
#对序列进行填充处理
input_array=tf.keras.preprocessing.sequence.pad_sequences(input,
           maxlen=400, truncating='post')
#使用模型获取转换后向量
output_array=model.predict(input_array)
print("输出矩阵的形状",output_array.shape)
```

```python
import tensorflow as tf      #导入所需库

# 构建RNN模型
model = tf.keras.Sequential()
model.add(tf.keras.layers.Embedding(10, 5, input_length=6))  # 嵌入层
model.add(tf.keras.layers.SimpleRNN(128))                    # 简单循环层
model.add(tf.keras.layers.Dense(5, activation='softmax'))    # 输出层
model.build(input_shape=(None,6))
model.summary()  # 显示网络模型参数信息
```
