```python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
# 加载数据
(x_train, y_train),(_,_) =  tf.keras.datasets.boston_housing.load_data(test_split=0)
print(x_train, y_train)
```

```python
# 查看数据集特征值和标签值的类型
print("特征值：",type(x_train))
print("标签值：",type(y_train))
```

```python
# 查看数据集特征值的属性
print("特征值维度：",x_train.ndim)
print("特征值的形状：",x_train.shape)
print("特征值元素的类型：",x_train.dtype)
```

```python
# 查看数据集标签值的属性
print("标签值维度:",y_train. ndim)
print("标签值的形状:",y_train. shape)
print("标签值元素的类型:",y_train. dtype)
```

```python
# 显示数据集中部分数据值
print("特征值最后6行数据：" )
print(x_train[-6:])
print("标签值最后6个数据：")
print(y_train[-6:])
```

```python
# 设置图形中的中文黑体
plt.rcParams['font.sans-serif']=['SimHei']
# 初始化列表
titles=["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
# 创建画布设置大小
plt.figure(figsize=(16,16))
```

```python
# 绘制子图
for i in range (13):
    plt.subplot(4,4, (i+1))
    plt.scatter(x_train[:,i],y_train)
    plt.xlabel(titles[i])
    plt.ylabel("Price($1000’s)")
    plt.title(str(i+1)+"."+titles[i]+"-Price")
# 自动调整子图
plt.tight_layout()
plt.show()
```
