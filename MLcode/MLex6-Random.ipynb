```python
# 导入模块
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
# 随机生成数据，数据格式为（x1,x2,y），x1、x2为数据的特征，y是数据所属类别。
X, y = make_blobs (n_samples=40,centers=2,random_state=6) # 生成40个线性可分的样本数据，分为2个中心。
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis',marker='o',edgecolor='k',s=100) # x轴，y轴，目标c
plt.title('Visualization of Generated Data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.show()
```

```python
# 训练SVM
# SVC是SVM的一种，用于分类
clf = svm.SVC(kernel='linear',C=1000)
clf.fit(X, y) 
print("参数W:",clf.coef_) # 斜率
print("参数b:",clf.intercept_)# 截距
print("支持向量：",clf.support_vectors_)
```

```python
# 预测
xpredict=[10,-10]
xpredict = np.array(xpredict).reshape(1,-1)
print(xpredict,"预测为：",clf.predict(xpredict),"类别")
xpredict =[10,-2]
xpredict = np.array(xpredict).reshape(1,-1)
print(xpredict,"预测为：",clf.predict(xpredict),"类别")
```

```python
from sklearn.datasets import make_moons
# 生成线性不可分的数据使用，make_moons生成线性不可分的数据，noise参数增加了数据的复杂性。
X,y=make_moons(n_samples=100,noise=0.1,random_state=6)
# 添加一个随机的第三维度
Z=np.random.rand(X.shape[0])
    # 随机生成第三维数据
X_3d=np.column_stack((X,Z))
    # 合并成三维数据
# 可视化三维数据点
fig=plt.figure(figsize=(14,8))
ax=fig.add_subplot(111,projection='3d')
scatter =ax.scatter(X_3d[:,0],X_3d[:,1],X_3d[:,2],c=y,cmap=plt.cm.Paired,alpha=0.7)
ax.set_xlabel('x1',fontsize=10)
ax.set_ylabel('x2',fontsize=10)
ax.set_zlabel('x3',fontsize=10)
# 调整视角
ax.view_init(elev=40,azim=40)
plt.show()
```

```python
#训练支持向量机模型，使用RBF核
clf = svm.SVC(kernel='rbf', C=1)
clf.fit(X_3d, y)
#可视化三维数据点
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_3d[:,0], X_3d[:,1], X_3d[:,2], c=y, cmap=plt.cm.Paired, alpha=0.7)
#绘制决策边界（超平面）
xx, yy = np.meshgrid(np.linspace(X_3d[:,0].min(), X_3d[:,0].max(), 20),
    np.linspace(X_3d[:,1].min(), X_3d[:,1].max(), 20))
ZZ = np.zeros(xx.shape)
#计算决策函数值
for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        Z = np.array([[xx[i,j], yy[i,j], 0]])
        ZZ[i,j] = clf.decision_function(Z)
#绘制决策超平面
ax.contour3D(xx, yy, ZZ, levels=[0], colors='k', linewidths=2)
#设置图形标题和标签
ax.set_xlabel('x1', fontsize=10)
ax.set_ylabel('x2', fontsize=10)
ax.set_zlabel('x3', fontsize=10)
plt.show()
```
