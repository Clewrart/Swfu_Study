```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import f1_score
import numpy as np
```

```python
# 导入数据，分别为输入特征和标签
wine =load_wine()
X=wine.data
Y=wine.target
# 划分训练集和测试集28
X_train,X_test,y_train,y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
# 输出结果
print("训练集特征形状：",X_train.shape)
print("测试集特征形状：",X_test.shape)
print("训练集标签形状:",y_train.shape)
print("测试集标签形状：",y_test.shape)
```

```python
# 随机森林模型训练和参数优化
rf = RandomForestClassifier(random_state=42)
#定义参数网格
param_grid = {
    'n_estimators':[50, 100, 200],
    'max_depth':[None, 5, 10],
    'min_samples_split':[2, 5, 10]
}
#执行网格搜索
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='accuracy', cv=3)
grid_search.fit(X_train, y_train)
#输出最佳参数
best_params = grid_search.best_params_
print("最佳参数:",best_params)
```

```python
# 使用最佳参数训练模型
best_model= grid_search.best_estimator_
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)

# 打印混淆矩阵
cm = confusion_matrix(y_test, y_pred)
print("混淆矩阵：")
print(cm)

# 可视化混淆矩阵
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',xticklabels=wine.target_names,yticklabels=wine.target_names)
plt.title('Confusion Matrix')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```

```python
# 特征重要性排序
# 获取特征重要性
importances = best_model.feature_importances_
if isinstance(X, pd.DataFrame):
    feature_names = X.columns
else:
    feature_names = [f'feature_{i}' for i in range(X.shape[1])]

# 创建数据框以便排序
feature_importance_df = pd.DataFrame({'Feature':feature_names,'Importance':importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
                                      
# 可视化特征重要性
plt.figure(figsize=(10, 6))
sns.barplot (x='Importance', y='Feature',data=feature_importance_df)
plt.title('Feature Importance from Random Forest')
plt.xlabel('Importance')
plt.ylabel ('feature')
plt.show()
```

```python
# 特征选择
selector = SelectFromModel (best_model, threshold="mean", prefit=True)
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test) 
```

```python
# 对选择后的模型训练和评估
rf_selected = RandomForestClassifier(random_state=42)
rf_selected.fit(X_train_selected, y_train)
y_pred_selected = rf_selected. predict(X_test_selected)

#计算选择后的性能指标
accuracy_selected=accuracy_score(y_test, y_pred_selected)
f1_selected = f1_score(y_test, y_pred_selected, average='weighted')

print(classification_report(y_test, y_pred_selected))
print(f'f1选择性能：',f1_selected)
```

```python
# 绘制性能对比图
performance_metrics = ['Accuracy','F1 Score']
f1_original = f1_score(y_test, y_pred, average='weighted')
accuracy_original = accuracy_score(y_test, y_pred)
original_performance = [accuracy_original,f1_original]
selected_performance =[accuracy_selected,f1_selected]
x = np.arange(len(performance_metrics))  # x轴

# 创建条形图
width=0.35   #条宽
fig, ax =plt.subplots(figsize=(10,6))
bars1 = ax.bar(x - width/2, original_performance, width,label='Original Features')
bars2 = ax.bar(x + width/2, selected_performance, width,label='Selected Features')
#添加标签和标题
ax.set_ylabel('Scores')
ax.set_title('Classifier Performance Comparison Before and After Feature Selection')
ax.set_xticks(x)
ax.set_xticklabels(performance_metrics)
ax.legend()
#在条上显示数值
def add_value_labels(bars):
    for bar in bars:
        yval = bar.get_height ()
        ax.text(bar.get_x()+ bar.get_width()/2,yval, round(yval,2),ha='center',va='bottom')
add_value_labels (bars1)
add_value_labels (bars2)
plt.tight_layout()
plt.show()
```
