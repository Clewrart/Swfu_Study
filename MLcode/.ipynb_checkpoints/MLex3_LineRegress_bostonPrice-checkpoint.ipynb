```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# 加载数据
(x_train, y_train),(x_test, y_test)=tf.keras.datasets.boston_housing.load_data(test_split=0.3)
# 查看数据
print("训练集样本数:",x_train.shape[0])
print("测试集样本数:",x_test.shape[0])
```

```python
# 建立、训练模型
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
# R²值和均方误差预测
y_pred= model.predict(x_test)
mse=mean_squared_error(y_test, y_pred)
r2=r2_score(y_test, y_pred)
print(f'Mean Squared Error:{mse:.2f}')
print(f'RSqur(R2) Score: {r2:.2f}')
```

```python
# 可视化真实值与预测值
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(10,6))
plt.scatter(y_test,y_pred,alpha=0.7)
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],color='red', linewidth=2)
plt.xlabel('真实房价')
plt.ylabel('预测房价')
plt.title('真实房价 vs 预测房价')
plt.grid()
plt.show()
```

```python
# 残差分析
residuals=y_test - y_pred
plt.figure(figsize=(10,6))
sns.histplot(residuals, bins=30, kde=True)
plt.title('残差分布')
plt.xlabel('残差')
plt.ylabel('频数')
plt.axvline(0, color='red', linestyle='--')
plt.show()
```

```python
# 相关性热图将训练集数据转换为DataFrame，确保列名与特征数匹配
train_data=pd.DataFrame(x_train, columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])

# 目标变量计算相关性矩阵
train_data['MEDV']= y_train
correlation_matrix=train_data.corr()

# 提取与目标变量的相关性并排序
correlation_with_target = correlation_matrix['MEDV'].drop('MEDV') # 去掉目标变量本身
sorted_features=correlation_with_target.abs().sort_values(ascending=False).index # 按绝对值排序

# 根据排序后的特征重新排列相关性矩阵
sorted_correlation_matrix=correlation_matrix.loc[sorted_features,sorted_features]

plt.figure(figsize=(12,10))
sns.heatmap(sorted_correlation_matrix,annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('特征相关性热图')
plt.show()
```

```python
# 特征重要性分析
coefficients =model.coef_

# DataFrame来显示特征名称及其对应的系数
feature_importance = pd.DataFrame({'Feature': train_data.columns[:-1], 'Coefficient': coefficients}) # 排除目标变量

# 按系数绝对值排序
feature_importance['Absolute Coefficient'] = feature_importance['Coefficient'].abs()
sorted_features = feature_importance.sort_values(by='Absolute Coefficient',ascending=False)

print('特征重要性排序:')
print(sorted_features[['Feature','Coefficient']])


plt.figure(figsize=(12,6))
sns.barplot(x='Coefficient',y='Feature',data=sorted_features, palette='viridis', hue='Feature', legend=False)
plt.title('特征重要性排序')
plt.xlabel('系数值')
plt.ylabel('特征')
plt.grid()
plt.show()
```
