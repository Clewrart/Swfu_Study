```python
# 导入所需模块和数据
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering # 层次聚类模块
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler # 标准化

iris = load_iris()
X = iris.data  # 特征数据
y = iris.target  # 实际标签
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
# 画树
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)# 标准化数据
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize=(12,9))
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
counta=int(len(X_scaled))# 计数
Z = linkage(X_scaled[ :counta], method='ward', metric='euclidean')

dendrogram(Z, labels=[f"Sample {i}" for i in range(counta)], 
           leaf_rotation=90, leaf_font_size=10)
plt.title('层次聚类树状图')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.axhline(y=3, color='r', linestyle='--')
plt.axhline(y=5, color='r', linestyle='--')
plt.axhline(y=10, color='r', linestyle='--')
plt.axhline(y=13, color='r', linestyle='--')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 进行层次聚类
hi = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hi = hi.fit_predict(X_scaled)
```

```python
# 评价
import seaborn as sns
from sklearn.metrics import confusion_matrix
confusion_mat = confusion_matrix(y,y_hi)

plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cluster 0', 'Cluster 1', 'Cluster 2'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Clusters')
plt.ylabel('True Classes')
plt.show()
```
