```python
# 导入必要的库
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
# 加载数据集
iris = load_iris()
features = iris.data
labels = iris.target
print(iris)
print(features)
print(labels)
```

```python
# 查看特征值和标签值的类型
print("特征值类型:", type(features))
print("标签值类型:", type(labels))
```

```python
# 查看特征值的属性
print("维度:", features.ndim)
print("形状:", features.shape)
print("元素数据类型:", features.dtype)
print("元素总个数:", features.size)
```

```python
# 查看标签值的属性
print("维度:", labels.ndim)
print("形状:", labels.shape)
print("元素数据类型:", labels.dtype)
print("元素总个数:", labels.size)
```

```python
# 显示前 10 条数据
print("前 10 条特征值")
print(features[:10])
print("前 10 条标签值数据")
print(labels[:10])
```

```python
# 绘制散点图
plt.rcParams['font.sans-serif']=['SimHei']
petal_length = features[:, 2]
petal_width = features[:, 3]
from matplotlib.colors import ListedColormap
cmap = ListedColormap(['blue', 'red', 'green'])
plt.figure(figsize=(8,6)) 
scatter = plt.scatter(petal_length, petal_width, c=labels, cmap=cmap)
plt.legend(handles=scatter.legend_elements()[0],labels=['Setosa(蓝色)', 'Versicolor(红色)', 'Virginica(绿色)'], title="品种")
plt.xlabel("长度")
plt.ylabel("宽度")
plt.title("鸢尾花散点图")
plt.tight_layout()
plt.show()
```
