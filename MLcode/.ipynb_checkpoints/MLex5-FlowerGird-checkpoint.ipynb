```python
# 导库文件及数据集
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.feature_selection import SelectFromModel
import xgboost as xgb
import lightgbm as lgb

# 加载鸢尾花数据集
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
iris_df = pd.DataFrame(X, columns=feature_names)
iris_df['target'] = y
iris_df['target_name'] = iris_df['target'].map({i: name for i, name in enumerate(target_names)})
print("数据集基本信息:")
print(f"特征数量: {X.shape[1]}")
print(f"样本数量: {X.shape[0]}")
print("\n特征名称:", feature_names)
print("\n类别名称:", target_names)
print("\n前5行数据:")
print(iris_df.head())

# 可视化特征之间的关系
plt.figure(figsize=(15, 10))
plt.rcParams['font.sans-serif']=['SimHei']
plt.suptitle("鸢尾花数据集特征可视化", y=1.02)

# 特征分布直方图
plt.subplot(2, 2, 1)
sns.histplot(data=iris_df, x='sepal length (cm)', hue='target_name', element='step', kde=True)
plt.title('花萼长度分布')

plt.subplot(2, 2, 2)
sns.histplot(data=iris_df, x='sepal width (cm)', hue='target_name', element='step', kde=True)
plt.title('花萼宽度分布')

plt.subplot(2, 2, 3)
sns.histplot(data=iris_df, x='petal length (cm)', hue='target_name', element='step', kde=True)
plt.title('花瓣长度分布')

plt.subplot(2, 2, 4)
sns.histplot(data=iris_df, x='petal width (cm)', hue='target_name', element='step', kde=True)
plt.title('花瓣宽度分布')

plt.tight_layout()
plt.show()

# 特征间关系散点图矩阵
sns.pairplot(iris_df, hue='target_name', diag_kind='kde')
plt.suptitle("特征间关系散点图矩阵", y=1.02)
plt.show()
```

```python
# 划分训练集和测试集七三比
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\n训练集大小: {X_train.shape[0]} 样本")
print(f"测试集大小: {X_test.shape[0]} 样本")
```

```python
# XGBoost和LightGBM分类
xgb_clf = xgb.XGBClassifier(random_state=42, eval_metric='mlogloss')
lgb_clf = lgb.LGBMClassifier(random_state=42)

xgb_params = {
    'n_estimators': [100], 
    'max_depth': [3, 5],
    'learning_rate': [0.1], 
    'subsample': [0.8],
    'colsample_bytree': [0.8] 
}

lgb_params = {
    'n_estimators': [100], 
    'max_depth': [3, 5], 
    'learning_rate': [0.1], 
    'num_leaves': [31] 
}

# 参数优化
xgb_grid = GridSearchCV(xgb_clf, xgb_params, cv=5, scoring='accuracy', n_jobs=-1)
xgb_grid.fit(X_train, y_train)
best_xgb = xgb_grid.best_estimator_
print("XGBoost最佳参数:", xgb_grid.best_params_)
print("XGBoost最佳交叉验证准确率:", xgb_grid.best_score_)

lgb_grid = GridSearchCV(lgb_clf, lgb_params, cv=5, scoring='accuracy', n_jobs=-1)
lgb_grid.fit(X_train, y_train)
best_lgb = lgb_grid.best_estimator_
print("LightGBM最佳参数:", lgb_grid.best_params_)
print("LightGBM最佳交叉验证准确率:", lgb_grid.best_score_)
```
